```python
import json
from pprint import pprint
json_data=open('data/2009-2010.json')

data_2009_2010 = json.load(json_data)
#print data_2009_2010
```

```python

#creates a dictionary of debaters with lists of all speaker points they have received
def create_dict_from_json(data) : 
    d = {}
    for i in data : 
        for j in i['debaters']: 
            if (d.has_key(j['ID'])):
                d[j['ID']].append(j['speaker points'])
            else:
                d[j['ID']] = [j['speaker points']]
    return d
dict_2009_2010 = create_dict_from_json(data_2009_2010)
#dict_2009_2010
    

        
```

```python
#the mean speaker points for a debater given the debater id and the speakers dict
import numpy as np


def calculate_debater_stats (debater,d) : 
    speaks = [float(x) for x in d[debater]]
    return np.mean(speaks),np.std(speaks)
            
calculate_debater_stats('23087',dict_2009_2010)
```

```python
#creates a dictionary of judges with the debater id and speaker points given for every round they've judged
def create_jdict_from_json (data) : 
    d = {}
    for i in data : 
        points_awarded = []
        for j in i['debaters'] : 
            points_awarded.append((j['ID'],j['speaker points']))
        if d.has_key(i['judge']['ID']) : 
            d[i['judge']['ID']] += points_awarded
        else : 
            d[i['judge']['ID']] = points_awarded
    return d 
judge_dictionary = create_jdict_from_json(data_2009_2010)

judge_dictionary
```

```python
#calculate the mean speaker points given by judge

def calculate_judge_stats(ID, d):
    speaks = [float(x) for _,x in d[ID]]
    return np.mean(speaks), np.std(speaks) 
calculate_judge_stats('1009', judge_dictionary)
```

```python
#looks up a debater's school, returns None if the debater is not in the data
def find_debater_school (ID,data) : 
    for i in data : 
        for j in i['debaters'] : 
            if j['ID'] == ID : 
                return j['school']
find_debater_school('27381', data_2009_2010)          
```

```python
#looks up a judge's school, return None if the judge is not in the data
def find_judge_school(ID,data) : 
    for i in data : 
        if i['judge']['ID'] == ID : 
            return i['judge']['school']
j = find_judge_school('10',data_2009_2010)
print j 
```

```python
#returns a list of speaker points awarded to a given debater by a given judge, returns an empty list if the judge or debater is not in the data
def get_speaker_points (judge,debater,data) : 
    lst = []
    for i in data : 
        if i['judge']['ID'] == judge : 
            for j in i['debaters'] : 
                if j['ID'] == debater : 
                    lst.append(j['speaker points'])
    return lst

get_speaker_points ('1033','23449',data_2009_2010)
```

```python
#returns a list of all the judge id's 
import sets
def judge_ids (data) : 
    lst = sets.Set()
    for i in data : 
        lst.add(i['judge']['ID'])
    return list(lst)
list_of_judge_ids = judge_ids (data_2009_2010)
 
```

```python

#returns a list of all the debater id's 
def debater_ids (data) : 
    lst = sets.Set()
    for i in data : 
        for j in i['debaters'] : 
            lst.add(j['ID'])
    return list(lst)
list_of_debater_ids = debater_ids (data_2009_2010)
```

```python
def get_tournament_numbers (j,data) : 
    tourns = []
    for i in data : 
        if (i['judge']['ID'] == j) : 
            tourns.append(i['tournament'])
    return tourns
get_tournament_numbers ('1033',data_2009_2010)
```

```python
#returns a dictionary for a given judge with their average score for every debater
def get_judge_averages (judge,data) : 
    d = {}
    for debater in debater_ids(data) : 
        scores = get_speaker_points(judge,debater,data) 
        float_scores = [float(s) for s in scores]
        if not(float_scores == []) : 
            d[debater] = np.mean(float_scores)
    return d
get_judge_averages ('1033',data_2009_2010)
```

```python
import matplotlib.pyplot as plt
def plot_judge_means (data): 
    means = []
    ids = judge_ids(data)
    d = create_jdict_from_json(data)
    for j in ids : 
        m,_ = calculate_judge_stats(j,d)
        means.append(m)
    plt.hist(means,bins=len(set(means)), log=False) 
    plt.ylabel('Count')
    plt.xlabel('Average Points')
    plt.title('Average Speaker Points Given by Judges')
    remove_border()
    plt.show()
    
#As you can see, when we first plot our data there are many abnormalities that prevent 
#us from seeing getting a good idea of the distribution of the data.
plot_judge_means(data_2009_2010)        



```

```python
def plot_debater_means (data) :
    means = []
    ids = debater_ids(data)
    d = create_dict_from_json(data)
    for i in ids:
        m,_ = calculate_debater_stats(i, d)
        means.append(m)
    plt.hist(means,bins=len(set(means)), log=False) 
    plt.ylabel('Count')
    plt.xlabel('Average Points')
    plt.title('Average Speaker Points Given to Debaters')
    remove_border()
    plt.show()
plot_debater_means(data_2009_2010)
    

```

```python
#a function that returns True if the data is good (no speaker points less than zero or greater than 30,
#no rounds where every debater has 0 speaker points, no data from tournament 103 (speaker points on a scale from 1-10)
def bad_data (d) : 
    ret = True
    all_zeros = True 
    if d['tournament'] == '103' : 
        ret = False
    for j in d['debaters'] : 
        speaks = float(j['speaker points'])
        if speaks > 0. :
            all_zeros = False
        ret &= (speaks >= 0. and speaks <= 30.)
    ret &= not all_zeros
    return ret
```

```python
#filters out bad data  
def remove_bad_data(data):
    return filter(bad_data,data)
revised_data_2009_2010 = remove_bad_data(data_2009_2010)
```
