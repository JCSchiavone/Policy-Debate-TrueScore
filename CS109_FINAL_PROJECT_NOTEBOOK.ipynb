```python
#Creates a dataframe from our json
def load_df(data):
    
    d = dict()
    
    d['tournament'] = [r['tournament'] for r in data]
    
    d['judge ID'] = [r['judge']['ID'] for r in data]
    d['judge school'] = [r['judge']['school'] for r in data]
    
    d['aff1 ID'] = [r['debaters'][0]['ID'] for r in data]
    d['aff2 ID'] = [r['debaters'][1]['ID'] for r in data]
    d['neg1 ID'] = [r['debaters'][2]['ID'] for r in data]
    d['neg2 ID'] = [r['debaters'][3]['ID'] for r in data]
    d['aff1 school'] = [r['debaters'][0]['school'] for r in data]
    d['aff2 school'] = [r['debaters'][1]['school'] for r in data]
    d['neg1 school'] = [r['debaters'][2]['school'] for r in data]
    d['neg2 school'] = [r['debaters'][3]['school'] for r in data]
    d['aff1 speaker points'] = [float(r['debaters'][0]['speaker points']) for r in data]
    d['aff2 speaker points'] = [float(r['debaters'][1]['speaker points']) for r in data]
    d['neg1 speaker points'] = [float(r['debaters'][2]['speaker points']) for r in data]
    d['neg2 speaker points'] = [float(r['debaters'][3]['speaker points']) for r in data]
    
    d['win'] = [int(r['win']) for r in data]
    
    df = pd.DataFrame.from_dict(d)
    df.index = range(len(data))
    
    return df
```

## Model

Our model, in its current iteration, is simply:

$$
D_i \sim Normal(\mu_i^{deb},\sigma_i^{deb}) \\
J_j \sim Normal(\mu_j^{jud},\sigma_j^{jud}) \\
Y_ij = D_i + J_j
$$

For each debater $i$, there is a normal distribution $D_i$ with mean $\mu_i^{deb}$ and standard deviation $\sigma_i^{deb}$. Similarly, for each judge $j$, there is a normal distribution $J_j$ with mean $\mu_j^{jud}$ and standard deviation $\sigma_j^{jud}$. 


```python
#DANIEL I SHOULD HAVE MENTIONED. WE ONLY WANT TO DO ONE DRAW WITH MET-HASTINGS, AND REJECT OR ACCEPT.
#SO ITERNUM SHOULD BE 1 AND I DON'T THINK WE WANT A BURN-IN PERIOD. 

# now obsolete - included in factor_gibbs

"""
Function
--------
sig2_d_draw

Uses Metropolis-Hastings algorithm to draw samples of sig2_d.

Inputs
--------
initial: initial guess for Metropolis-Hastings algorithm, default=1.0
iternum: number of iterations/steps for algorithm, default=1000
stepsize: step size, variance of Gaussians used in algorithm, default=2.0
burn-in: length of burn-in period, default=200
thinning: 1/thinning is fraction of values to keep, default=1

Outputs
--------
List of samples of sig2_d

"""
def sig2_d_draw(initial=1.0, iternum=1000, stepsize=2.0, burn_in=200, thinning=1):
    # initialize list to store samples
    samples = [initial] 

    accepted = 0.0

    for i in range(iternum):
        x_old = samples[len(samples)-1]
        x_new = np.random.normal(x_old, stepsize)
        
        # calculate acceptance ratio -- NOT DONE -- REQUIRES POSTERIOR
        if p(x_new)/p(x_old) >= 1:
            samples.append(x_new)
            accepted += 1
        else:
            randnum = np.random.uniform(0.0,1.0)
            if x_new >= randnum:
                samples.append(x_new)
                accepted += 1
            else: 
                samples.append(x_old)

    # acceptance ratio should ideally be ~45%
    print "Acceptance Ratio: %f" % accepted/iternum
    
    # account for burn-in and thinning as necessary
    samples_clean = []
    for j in range(burn_in, iternum):
        if j % thinning == 0:
            samples_clean.append(samples[j])
    
    return samples_clean
```

```python
"""
Function
--------
mu_d_draw

Draw a single sample from the conditional posterior distribution of mu_m.

Inputs
--------
X_d: Matrix of dimensions g-by-L+1, in current case, L=0, g=# of ratings for d
Y_d: Vector of length g
mu_d_sig2: Variance for d
mu_j_sig2: Variance for j

Outputs
--------
Single draw from conditional posterior

"""
def mu_d_draw(X_d, Y_d, mu_d_sig2, mu_j_sig2):
    #Compute matrices that define conditional posterior.
    Q_m_inv = np.linalg.inv(np.dot(X_d.T, X_d)/(mu_d_sig2 + mu_j_sig2)+1./mu_d_sig2)
    XtY = np.dot(X_d.T, Y_d)

    #Draw item-specific parameters.
    return np.random.multivariate_normal(np.dot(Q_m_inv, XtY)/(mu_d_sig2 + mu_j_sig2), Q_m_inv)
```

```python
"""
Function
--------
mu_j_draw

Draw a single sample from the conditional posterior distribution of mu_j.

Inputs
--------
X_j: Matrix of dimensions g-by-L+1, in current case, L=0, g=# of ratings for d
Y_j: Vector of length g
mu_d_sig2: Variance for d
mu_j_sig2: Variance for j

Outputs
--------
Single draw from conditional posterior

"""
def mu_j_draw(X_j, Y_j, sig2_d, sig2_j):
    #Compute matrices that define conditional posterior.
    Q_m_inv = np.linalg.inv(np.dot(X_j.T, X_j)/(mu_d_sig2 + mu_j_sig2)+1./mu_j_sig2)
    XtY = np.dot(X_j.T, Y_j)

    #Draw item-specific parameters.
    return np.random.multivariate_normal(np.dot(Q_m_inv, XtY)/(mu_d_sig2 + mu_j_sig2), Q_m_inv)
```

```python
"""    
Function
--------
factor_gibbs

Runs a gibbs sampler to infer the judge-specific and debater-specific factors

Inputs
-------
data: A dataframe containing ratings data.
maxit: Number of samples to draw from posterior.
mu_d_sig2: Hyperparameter controlling regularization of mu_d
mu_j_sig2: Hyperparameter controlling regularization of mu_j
v_d: Parameter for distribution of sig2_d
v_j: Parameter for distribution of sig2_j
mhstep: Metropolis-Hastings algorithm stepsize
progress: if true, print iteration number every 100 iterations.

Outputs
--------
Dictionary with elements
mu_d: Draws of mu_d. D-by-1-by-maxiter array.
mu_j: Draws of mu_j. J-by-1-by-maxiter array.
sig2_d: Draws of sig_d,  D-by-1-by-maxiter array
sig2_j: Draws of sig_j,  J-by-1-by-maxiter array
EY: Draws of fitted values of Y. N-by-maxiter array.

"""
def factor_gibbs(data, maxit, mu_d_sig2, mu_j_sig2, v_d, v_j, mhstep, progress=True):
    data = data.copy()
    # number of rows
    N = data.shape[0] 

    # number of interaction variables
    L = 0 

    #Initialize parameters and set up data structures for holding draws.
    #Mean
    mu_d_draws = np.zeros(maxit)
    mu_j_draws = np.zeros(maxit)
    
    #Variance
    sig2_d_draws = np.zeros(maxit)
    sig2_j_draws = np.zeros(maxit)

    # initialize mu_d, mu_j
    mu_d = np.eyes(maxit)*get_avg_speaks(data)
    mu_j = np.zeros(maxit)
    
    #Matrix for holding the expected number of stars for each observation at each draw from the posterior.
    EY_draws = np.zeros(maxit)
    
    dpoints = get_debater_scores(data)
    jpoints = get_judge_ratings(data)
    
    #Main sampler code
    for i in range(maxit):
        if i%100==0 and progress:
            print i

        #The entire regression equation except for the overall mean.
        EY = mu_j + mu_d + np.random.normal(0, sig2_d + sig2_j) # must index to correct subsets of the data
        
        #Compute the expectation of each observation given the current parameter values.
        EY_draws[i]=EY

        # Draw mu_d for each debater
        for d in dpoints:
            Y_d = dpoints - mu_d
            X_d = mu_j

            mu_d[d] = mu_d_draw(X_d, Y_d, mu_d_sig2, mu_j_sig2)

        # Draw mu_j for each judge
        for j in jpoints:
            Y_j = jpoints - mu_j
            X_j = mu_d

            mu_j[j] = mu_j_draw(X_j, Y_j, sig2_d, sig2_j)
            
        # Draw v_d using Metropolis-Hastings algorithm
        sig2_d_new = np.random.normal(sig2_d, mhstep)
        # calculate acceptance ratio -- NOT DONE -- REQUIRES POSTERIOR
        if p(sig2_d_new)/p(sig2_d) >= 1:
            sig2_d = sig2_d_new
        else:
            randnum = np.random.uniform(0.0,1.0)
            if sig2_d_new >= randnum:
                sig2_d = sig2_d_new    
                
        # Draw v_j using Metropolis-Hastings algorithm
        sig2_j_new = np.random.normal(sig2_j, mhstep)
        # calculate acceptance ratio -- NOT DONE -- REQUIRES POSTERIOR
        if p(sig2_j_new)/p(sig2_j) >= 1:
            sig2_j = sig2_j_new
        else:
            randnum = np.random.uniform(0.0,1.0)
            if sig2_j_new >= randnum:
                sig2_j = sig2_j_new

        #Record draws
        mu_d_draws[i] = mu_d
        mu_j_draws[i] = mu_j
        sig2_d_draws[i] = sig2_d
        sig2_j_draws[i] = sig2_j

    return {"mu_d": mu_d_draws, "mu_j": mu_j_draws, "v_d": v_d_draws, "v_j": v_j_draws, "EY": EY_draws}
```

## Validation

Our current method for validation is $\prod_{i,j} P({SP}_{|i,j}|Y_{i,j})$

Pseudo-code:
def validate(data, f, N)

* data is the dataset
* f is the model 

def f(data)

returns 4 vectors of the parameters ($\mu_i^{deb},\sigma_i^{deb}, \mu_j^{jud},\sigma_j^{jud}$)

* N is the number of holdouts - may need to change if computation is prohibitively expensive


```python
#takes in our data, a function to create a model, and the number of holdouts
#returns the root mean square error of the model 
import copy
import scipy.stats as sp
def validate (data,f,N) : 
    error = []
    
    #list of holdouts to test our model on 
    holdouts = random.sample(data,N) 
    for holdout in holdouts :
        new_data = copy.copy(data)
        
        #creates our model using all of the data without the holdout as training data
        new_data.remove(holdout)
        deb_mean,deb_sd,judge_mean,judge_sd = f(new_data)
        
        #computes the error between our prediction(based on our model) for the speaker points
        #awarded in the holdout round and the actual points awarded in the holdout round
        for i in holdout['debaters'] : 
            rv = sp.norm(loc = deb_mean[i['ID']] + judge_mean[holdout['judge']['ID']], 
                         scale = deb_sd[i['ID']] + judge_sd[holdout['judge']['ID']])
            error.append = rv.pdf(deb_mean[i['ID']] + judge_mean[holdout['judge']['ID']]) - rv.pdf(i['speaker points'])
   
    #returns the root mean square error of our predictions
    return np.sqrt(np.mean([e**2 for e in error]))
```
